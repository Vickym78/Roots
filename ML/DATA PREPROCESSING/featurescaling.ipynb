```python
# lets import dataset 
import pandas as pd 
import numpy as np
import seaborn as sns
```

TYPE OF FEATURE SCALING USING STANDARISATION AND MINMAX

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler
```

```python
# for mean and std use .fit()  to aplly trasform()

scaler.fit(df[['total_bill']])
```

```python
scaler.transform(df[['total_bill']])
```

IN THIS WAY WE DONE IT at once


```python
new_df = pd.DataFrame(scaler.fit_transform(df[['total_bill' ,'tip']]) , columns=['total_bill' , 'tip'] )

new_df
```
