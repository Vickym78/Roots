### Imbalance in data means there is huge difference between Majority and minority class


```python
# create database
import pandas as pd
import numpy as np

no_class_0=900
no_class_1=100
```

```python
class_0=pd.DataFrame({
    'f1':np.random.normal(0,1,no_class_0),
    'f2':np.random.normal(0,1,no_class_0),
    'target':[0]*no_class_0

})

class_0
```

```python
class_1=pd.DataFrame({
    'f1':np.random.normal(0,3,no_class_1),
    'f2':np.random.normal(0,3,no_class_1),
    'target':[1]*no_class_1

})

class_1
```

```python
df = pd.concat([class_0,class_1]).reset_index(drop=True)
df
```

```python
df.target.value_counts()
```

Doing with upsampling of of minority and make it till majority

```python
df_minority = df[df.target==1]
df_majority = df[df.target==0]
```

# do sampling with help of sklearn

```python
from sklearn.utils import resample
df_updated_minority = resample(df_minority ,replace=True , n_samples=len(df_majority), random_state=42)
df_updated_minority
```

```python
updated_df = pd.concat([df_majority,df_updated_minority])
updated_df
```

NOW THE NO OF MINORITY AND MAJORITY IS EQUAL

```python
updated_df.target.value_counts()
```

###     Now for downsampling majority class reduce to minority class

```python
df_updated_majority = resample(df_majority,replace=True , n_samples=len(df_minority), random_state=42)
df_updated_majority
```

```python
updated_df_2 = pd.concat([df_updated_majority,df_minority])
updated_df_2
```

```python
updated_df_2.target.value_counts()
```
